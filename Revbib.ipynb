```python
import pandas as pd
```

```python
file_name = "savedrecs.txt"
```

```python
with open("savedrecs.txt", encoding="utf-8-sig") as f:
    lines = list(f)
```

```python
abstracts = []
titles = []
for line in lines: 
    if line.startswith("AB "):
        abstracts.append(line[3:])
    if line.startswith("TI "):
        titles.append(line[3:])
```

```python
len(titles)
```

```python
df = pd.DataFrame({
    "title": titles,
    "abstract": abstracts
    
})
```

```python
df
```

```python
mask = df["abstract"].str.contains(
    r"\bethic\w*\b",
    case=False,
    na=False,
    regex=True
)

df_ethics = df[mask]
df_ethics ["abstract"].head()
```
